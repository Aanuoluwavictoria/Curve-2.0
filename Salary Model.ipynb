# Split the into traning and test:

Splinting the dataset into the tranig and traning set and testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/3, random_state = 0)
```

# Fit simple Linear Regression to the Traning dataset 

Fitting simple leaning regression to the traning dataset

```python
from sklearn.linear_model import LinearRegression
ir = LinearRegression()
ir.fit(x_train, y_train)
ir.fit(x_test,y_test)
```

# Predict test set results

```python
y_pred = ir.predict(x_test)
y_pred
```

```python
y_pred = ir.predict(x_train)
y_pred
```

# Visualize the train set results

```python
plt.scatter(x_train, y_train, color = 'blue')
plt.plot(x_train, ir.predict(x_train), color='red')
plt.title('Salary~Exprerience(Train set)')
plt.xlabel('Year of Experience')
plt.ylabel('Salery')
plt.show()
```

```python
plt.scatter(x_train, y_train, color = 'blue')
plt.plot(x_train, ir.predict(x_train), color='red')
plt.title('Salary vs Exprerience(Train set)')
plt.xlabel('Year of Experience')
plt.ylabel('Salery')
plt.show()
```
